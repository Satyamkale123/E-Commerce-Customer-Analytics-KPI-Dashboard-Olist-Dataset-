```python
import pandas as pd

df = pd.read_csv("olist_order_summary.csv")

df['order_date'] = pd.to_datetime(df['order_date'])
df['order_month'] = df['order_date'].dt.to_period('M')

# First purchase month for each customer
first_purchase = df.groupby('customer_id')['order_month'].min().reset_index()
first_purchase.columns = ['customer_id','first_purchase_month']

# Merge
df = df.merge(first_purchase, on='customer_id')

# Cohort index (months since first purchase)
df['cohort_index'] = (df['order_month'] - df['first_purchase_month']).apply(lambda x: x.n)

# Aggregation
cohort_table = df.groupby(['first_purchase_month', 'cohort_index'])['customer_id'].nunique().reset_index()

cohort_table.to_csv("olist_customer_cohort.csv", index=False)
```
